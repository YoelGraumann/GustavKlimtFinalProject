```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plot_barplot(df_column):
    """
    Function to plot a black-and-white barplot for a given column in a DataFrame.

    Parameters:
    df_column (pd.Series): The column from the DataFrame to plot.

    Returns:
    None
    """
    # Count the occurrences of each unique value in the column
    value_counts = df_column.value_counts().reset_index()
    value_counts.columns = ['Category', 'Count']

    # Set the style and color palette for black-and-white
    sns.set(style="whitegrid")
    sns.set_palette("Greys")

    # Create the barplot
    plt.figure(figsize=(8, 6))
    sns.barplot(x='Category', y='Count', data=value_counts, edgecolor='black')

    # Customize the plot
    plt.title(f"{df_column.name}", fontsize=16)
    plt.xlabel("", fontsize=12)
    plt.ylabel("Count", fontsize=12)

    # Remove the top and right spines for a cleaner look
    sns.despine()

    # Show the plot
    plt.tight_layout()
    plt.show()

# Example usage:
# df = pd.DataFrame({'ExampleColumn': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'A']})
plot_barplot(survey_results['Which group most closely matches or reflects the art style of Gustav Klimt?'])
```
